```python
#EXPLORATORY DATA ANALYSIS (EDA PHASE)
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y='Amount', data = sales_gen)
```

```python
#From Above Graphs we understand that moost of the buyers are females and even the purchasing power of females are greater than males
```

```python
ax = sns.countplot( data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

```python
# From Above graph we can understand that most of the buyers are of age between 26-35 Females
```

```python
#total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
#total number of amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state, x='State', y='Amount')
```

```python
# from above insigts we can understand that most of the orders and sales have been placed from Uttar Pradesh, Maharsahtra and Karnataka Respectively.
```

```python
# for marital status
msax = sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in msax.containers:
  msax.bar_label(bars)
```

```python
sales_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_status,x='Marital_Status',y='Amount', hue='Gender')
```

```python
#From above insights we understand that most of the purchase has been placed byb married women and the purchase power is also high of married women respectivley
```

```python
# OCCUUPATION based on orders
sns.set(rc={'figure.figsize':(20,5)})
oax = sns.countplot(x='Occupation', data = df)
for bars in oax.containers:
  oax.bar_label(bars)
```

```python
# OCCUPATION based on Amount/Sales
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation', y='Amount', data=sales_occ)
```

```python
#From abvove insight we can understand that most of the buyers are workin in IT, Healthcare and Aviation Sectors.
```

```python
# Product Category based on order
sns.set(rc={'figure.figsize':(30,5)})
pcax = sns.countplot(x='Product_Category', data=df)
for bars in pcax.containers:
  pcax.bar_label(bars)
```

```python
sales_catg=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sns.barplot(x='Product_Category',y='Amount', data=sales_catg)
```

```python
# From the above order we can see that most of the sales have placed from Food, Clothing and Electronic Category
```

```python
sales_catg=df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(7)
sns.barplot(x='Product_Category',y='Orders', data=sales_catg)
```

```python
# Now from the above insights it cleary visible that highest selling products are from Clothing, Food and Electronics sector.
```

Married Women of age group 26-35 from Uttar Pradesh, Maharashtra and Karnataka working in IT, Healthcare and Aviation sector are more likey to buy products from Food, Clothing and Electronics Category.
